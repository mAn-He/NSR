```python
def find_difference(folders, item_name):
   
    difference = list(set(folders) - set(item_name))
    return difference

result = sorted(find_difference(folders, item_name))
print(sorted(result))  
```

```python

# 이미지 폴더 경로
image_folder = '이미지'

# 이미지 폴더 내의 모든 폴더 이름을 변경
for old_name, new_name in name_mapping.items():
    old_path = os.path.join(image_folder, old_name)
    new_path = os.path.join(image_folder, new_name)
    try:
        if os.path.exists(old_path):
            os.rename(old_path, new_path)
            print(f"Renamed folder '{old_path}' to '{new_path}'")
        else:
            print(f"Folder '{old_path}' does not exist")
    except FileNotFoundError as e:
        print(f"Error renaming '{old_path}' to '{new_path}': {e}")
    except Exception as e:
        print(f"An unexpected error occurred while renaming '{old_path}' to '{new_path}': {e}")
```

```python
with open('name_mapping.txt', 'w') as file:
    for old_name, new_name in name_mapping.items():
        file.write(f"{old_name}: {new_name}\n")
```

```python
import os
import shutil

def move_black_images(source_folder):
    # 'BLACK' 폴더 경로 설정
    black_folder = os.path.join(source_folder, 'BLACK')
    
    # 'BLACK' 폴더가 없으면 생성
    if not os.path.exists(black_folder):
        os.makedirs(black_folder)
        print(f"Created directory: {black_folder}")
    
    # 소스 폴더 내 파일들을 검사
    for filename in os.listdir(source_folder):
        file_path = os.path.join(source_folder, filename)
        
        # 파일인지 확인하고, 파일명에 'BLACK'이 포함되어 있는지 확인
        if os.path.isfile(file_path) and 'BLACK' in filename.upper():
            # 파일을 'BLACK' 폴더로 이동
            shutil.move(file_path, black_folder)
            print(f"Moved file '{file_path}' to '{black_folder}'")
```

```python
for dir in os.listdir('이미지'):
    move_black_images(dir)
```

```python
def move_files_to_black_folder_if_no_black_files(source_folder):
    # 'BLACK' 폴더 경로 설정
    black_folder = os.path.join(source_folder, 'BLACK')

    # 폴더 내 'BLACK'으로 시작하는 파일이 있는지 확인
    black_files_exist = any(
        filename.startswith('BLACK') for filename in os.listdir(source_folder)
        if os.path.isfile(os.path.join(source_folder, filename))
    )
    
    # 'BLACK'으로 시작하는 파일이 없으면
    if not black_files_exist:
        # 'BLACK' 폴더가 없으면 생성
        if not os.path.exists(black_folder):
            os.makedirs(black_folder)
            print(f"Created directory: {black_folder}")
        
        # 소스 폴더 내 모든 파일을 'BLACK' 폴더로 이동
        for filename in os.listdir(source_folder):
            file_path = os.path.join(source_folder, filename)
            if os.path.isfile(file_path):
                shutil.move(file_path, black_folder)
                print(f"Moved file '{file_path}' to '{black_folder}'")
```

```python
move_files_to_black_folder_if_no_black_files(full_paths[2])
```

```python
for dir in full_paths:
    move_files_to_black_folder_if_no_black_files(dir)
```

```python
def organize_images_by_color(source_folder, colors):
    # 소스 폴더 내 파일들을 검사
    for filename in os.listdir(source_folder):
        file_path = os.path.join(source_folder, filename)
        
        # 파일인지 확인
        if os.path.isfile(file_path):
            # 파일명이 colors 리스트의 원소로 시작하는지 확인
            for color in colors:
                if filename.startswith(color):
                    # 해당 color 폴더 경로 설정
                    color_folder = os.path.join(source_folder, color)
                    
                    # 해당 color 폴더가 없으면 생성
                    if not os.path.exists(color_folder):
                        os.makedirs(color_folder)
                        print(f"Created directory: {color_folder}")
                    
                    # 파일을 해당 color 폴더로 이동
                    shutil.move(file_path, color_folder)
                    print(f"Moved file '{file_path}' to '{color_folder}'")
                    break  # 파일이 이동되었으면 더 이상 다른 색상을 확인할 필요 없음
```

```python
organize_images_by_color(full_paths[4],colors)
```

```python
def remove_spaces_around_slash(path):
    parts = path.split('/')
    # 각 부분의 앞뒤 공백 제거
    parts = [part.strip() for part in parts]
    # '/'로 다시 연결
    return '/'.join(parts)

# '파일경로' 열에 적용
nsr['파일경로'] = nsr['파일경로'].apply(remove_spaces_around_slash)
```

```python
def list_files_in_directory(directory):
    if os.path.exists(directory) and os.path.isdir(directory):
        # 파일 목록을 경로와 함께 반환
        return [os.path.join(directory, filename) for filename in os.listdir(directory)]
    else:
        return []
```
