```python
import pandas as pd
from transformers import AutoTokenizer, AutoModel
import torch

tokenizer = AutoTokenizer.from_pretrained("klue/roberta-base")
model = AutoModel.from_pretrained("klue/roberta-base")
```

```python
def text_to_vector(texts):
    # 텍스트 인코딩 (토크나이징 및 텐서 변환)
    inputs = tokenizer(texts, padding=True, truncation=True, max_length=512, return_tensors="pt")
    # 모델을 평가 모드로 설정
    model.eval()
    # forward pass, 그라디언트 계산 방지
    with torch.no_grad():
        outputs = model(**inputs)
    # [CLS] 토큰에 해당하는 벡터 추출 (각 텍스트의 첫 번째 토큰)
    vectors = outputs.last_hidden_state[:,0,:].detach().numpy()
    return vectors
```

```python
def vector_to_columns(df, column_prefix, vectors):
    for i in range(vectors.shape[1]):  # 벡터 차원만큼 반복
        df[f'{column_prefix}_{i}'] = vectors[:, i]
    return df



# 결과 확인
# print(nsr_ss1_sum.head())
```

```python
X = nsr_total_sum.drop(columns=['판매수량'])
y= nsr_total_sum['판매수량']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error
import numpy as np

# 가정: 'X'는 특성 데이터, 'y'는 타겟 데이터
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# KNN 회귀 모델 초기화 (기본적으로 이웃의 수는 5)
knn = KNeighborsRegressor(10)

# 모델 학습
knn.fit(X_train, y_train)

# 예측 수행
y_pred = knn.predict(X_test)

# RMSE 계산
rmse = np.sqrt(mean_squared_error(y_test, y_pred))

print(f"테스트 데이터에 대한 RMSE: {rmse}")
```
